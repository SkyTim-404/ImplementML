```python
import sys
import math
import matplotlib.pyplot as plt

sys.path.append(r"../")

from dataset import Dataset
from models import ANN
```

Load mnist dataset

```python
train = Dataset.get_train_dataset(r"../../data")
test = Dataset.get_test_dataset(r"../../data")

X_train, y_train = train.data.numpy(), train.targets.numpy()
X_test, y_test = test.data.numpy(), test.targets.numpy()
```

Visualize data

```python
plt.imshow(X_train[0], cmap=plt.cm.binary)
plt.show()
```

```python
num_to_display = 16
num_images_per_row = math.ceil(math.sqrt(num_to_display))

plt.figure(figsize=(10, 10))

for i in range(num_to_display):
    # num of rows, num of columns, index (start from 1)
    plt.subplot(num_images_per_row, num_images_per_row, i+1)
    plt.xticks([])  #remove x axis
    plt.yticks([])  #remove y axis
    plt.grid(False) #remove grid lines
    plt.imshow(X_train[i], cmap=plt.cm.binary)
    plt.xlabel(y_train[i])  #label with correct labels

plt.show()
    
```

Hyperparameters

```python
learning_rate = 0.001
epoch = 1

load_model = True
```

Load model

```python
model = ANN()

if load_model:
    model.load()
```

Train model

```python
model.train(X_test, y_test, learning_rate, epoch)
model.save()
```

```python
y_pred = model.predict(X_test)
print(f"First 10 predictions: \t{y_pred[:10]}")
print(f"First 10 actual: \t{y_test[:10]}")
accuracy = model.score(X_test, y_test)
print(f"Accuracy: {accuracy}")
```
